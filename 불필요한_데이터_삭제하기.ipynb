**데이터 불러오기(헤더파일)**

```python
import gdown as gd
gd.download('https://bit.ly/3RhoNho','ns_202104.csv',quiet=False)
```

```python
import pandas as pd
ns_df = pd.read_csv('ns_202104.csv',low_memory=False)
ns_df.head()
```

열 삭제하기 - loc

```python
ns_book = ns_df.loc[:,'번호':'등록일자'] #loc=열
ns_book.head()
```

```python
#columns (행) / 기준이 없으면 첫번째 불러옴 / 원본파일을 불러왔따->원본파일은 왠만하면 수정하지 말자
print(ns_df.columns)
```

```python
#인덱스 클래스 : 인덱스에 있는 열을 돌면서 비교함
selected_columns = ns_df.columns != 'Unnamed: 13'
```

```python
#True 인 열의 모든 행을 선택
ns_book1 = ns_df.loc[:,selected_columns]
ns_book1.head()
```

```python
#인덱스 클래스 : 인덱스에 있는 열을 돌면서 비교함
selected_columns = ns_df.columns != '부가기호'
#True 인 열의 모든 행을 선택
ns_book = ns_df.loc[:,selected_columns]
ns_book.head()
```

```python
'''drop 매서드 : 헹 또는 열 삭제
axis == 0 >행 삭제
axis == 1 >열 삭제'''
```

```python
ns_book1 = ns_df.drop('Unnamed: 13',axis=1)
ns_book1.head()
```

```python
#여러개의 열 삭제
ns_book1 = ns_df.drop(['Unnamed: 13','부가기호'], axis=1)
ns_book1.head()
```

```python
#여러개의 열 삭제
ns_book1 = ns_df.drop(['저자'], axis=1,inplace=True)
ns_book1.head()
```

```python
ns_book1 = ns_df.dropna(axis=1)
ns_book1.head()
#pandas에서는 비어있는 값을 Nan 으로 표시하는데 dropna()메서드는 nan 이 하나라도 포함된 열인 행을 기준으로 다 삭제해버림
```

```python
#모든 값이 NaN인 열을 삭제
ns_book1 = ns_df.dropna(axis=1, how='all')
ns_book1.head()
```

행 삭제하기

```python
#인덱스 0~1까지 2개의 행 삭제
ns_book2 = ns_book.drop([0,1])
ns_book2.head()
```

```python
#인덱스가 0~2까지를 제외한 나머지 선택
ns_book2 = ns_book[3:]
ns_book2.head()
```

```python
#인덱스가 0~1까지 선택
ns_book2 = ns_book[:2]
ns_book2.head()
```

```python
#불리언 배열 활용 - 비교 연산자를 활용하여 원하는 행은 True 제외할 행은 False로 표시
##출판사가 한빛미디어 인 행 만 추출
seleceted_rows = ns_df['출판사'] == '한빛미디어'
ns_book2 = ns_book[seleceted_rows]
ns_book2.head()
```
